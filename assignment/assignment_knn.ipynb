**Q0.**
1. What is the difference between regression and classification?
2. What is a confusion table? What does it help us understand about a model's performance?
3. What does the SSE quantify about a particular model?
4. What are overfitting and underfitting? 
5. Why does splitting the data into training and testing sets, and choosing $k$ by evaluating accuracy or SSE on the test set, improve model performance?
6. With classification, we can report a class label as a prediction or a probability distribution over class labels. Please explain the strengths and weaknesses of each approach.

My Answers:
1) Regression is used to predict numeric values while classification is used to predict categorization of values. Typically, regression is used for continuous values and establishes a relationship between variables. Classification predicts categories for discrete values and is primarily responsible for assigning data to categories.
2) A confusion table is used to evaluate the classification model's performance. It is a summarized table of the number of both correct and incorrect predictions and is broken down by each class. The confusion table or matrix breaks down the number of True Positives, True Negatives, False Positive, and False Negative. The 'True' values refer to the values accurately predicted by the model, and the 'False' refer to values incorrectly classified by the model.
3) The SSE calculates the amount of variation in the data by a model by calculating the difference between the predicted values and the actual data.
4) Overfitting refers to when the model predicts values to close to the original data set. The model is essentially more so memorizing the values from the original data set instead of detecting the pattern and being able to generalize it to other datasets. Underfitting is when the data is too generalized and leaves out some specificities or smaller patterns as a result of smoothing the curve out too much.
5) Splitting the data into training and testing sets ensures that the model learns the general patterns within the data rather than the specific and unique identifiers or noise within the data. Choosing k based off the SSE test ensures that we are choosing the best value that ensures variance and accuracy while balancing bias. 
6) Reporting a class label is a more direct and clear cut method which also requires less computational resources since it outputs a single class. It directly indicates the predicted class. However, it does have limited insight since it does not display how close one prediction was to being categorized as a different class which may be useful for improving the model's accuracy. Reporting a probability distribution over class labels is useful for providing a direct confidence level for each class prediction. However, it can be computationally demanding to process multiple probability distributions can be more intensive to output compared to a single class output.

**Q6.** This is a case study on $k$ nearest neighbor regression and imputation, using the `airbnb_hw.csv` data.

There are 30,478 observations, but only 22,155 ratings. We're going to build a kNN regressor to impute missing values. This is a common task, and illustrates one way you can use kNN in the future even when you have more advanced models available.

1. Load the `airbnb_hw.csv` data with Pandas. We're only going to use `Review Scores Rating`, `Price`, and `Beds`, so use `.loc` to reduce the dataframe to those variables.
2. Set use `.isnull()` to select the subset of the dataframe with missing review values. Set those aside in a different dataframe. We'll make predictions about them later.
3. Use `df = df.dropna(axis = 0, how = 'any')` to eliminate any observations with missing values/NA's from the dataframe.
4. For the complete cases, create a $k$-NN model that uses the variables `Price` and `Beds` to predict `Review Scores Rating`. How do you choose $k$? (Hint: Train/test split, iterate over reasonable values of $k$ and find a value that minimizes SSE on the test split using predictions from the training set.)
5. Predict the missing ratings. 
6. Do a kernel density plot of the training ratings and the predicted missing ratings. Do they look similar or not? Explain why.

```python
# setting up environment
import numpy as np
import pandas as pd
import os
import seaborn as sns
import matplotlib as plt
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import StandardScaler
```

```python
# load in data and preview:
airbnb_df = pd.read_csv('../data/airbnb_hw.csv')
print(airbnb_df.columns)
airbnb_df.head(5)
```

```python
airbnb_df['Price'].unique()
# cleaning price to remove commas:
airbnb_df["Price"] = airbnb_df["Price"].replace({",": "", "\\$": ""}, regex=True).astype(float)
```

```python
# only keeping columns 'Review Scores Rating', 'Price', 'Beds':
rm_airbnb = airbnb_df.loc[:, ['Review Scores Rating', 'Price', 'Beds']]
rm_airbnb.head(5)
```

```python
# missing review values:
df_missing_reviews = rm_airbnb.isnull()
df_missing_reviews.head(5)
```

```python
# eliminating observations with missing values
df_full_reviews = rm_airbnb.dropna(axis = 0, how = 'any')
df_full_reviews.head(10)
```

```python
# create target variable vectors
y = df_full_reviews["Review Scores Rating"]
X = df_full_reviews.loc[:, ["Price", "Beds"]] # creating feature matrix

# normalizing X
X = (X - X.min()) / (X.max() - X.min())

k_bar = 20 # setting number of k to try

for _ in range(5): 
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

    SSE = []  # storing SSE values
    for k in range(k_bar):
        model = KNeighborsRegressor(n_neighbors=k + 1)
        fitted_model = model.fit(X_train, y_train) # training model on our data
        y_hat = fitted_model.predict(X_test) # predicting values for test set
        SSE.append(np.sum((y_test - y_hat) ** 2)) # appending the SSE list with the computed SSE values for test set

    # finding the lowest recorded SSE
    SSE_min = np.min(SSE)
    min_index = np.where(SSE == SSE_min) # finding the indices of SSE that are equal to the min
    k_star = min_index[0][0] + 1 
    print(f"Optimal k: {k_star}")

    # Plot SSE by k
    plt.plot(np.arange(1, k_bar + 1), SSE, label="Test SSE")
    plt.xlabel("k")
    plt.ylabel("SSE")
    plt.title("SSE")
```

```python
# Initialize StandardScaler
scaler = StandardScaler()
df_full_reviews = df_full_reviews.dropna(axis = 0, how = 'any')

# standardizing Price columns:
df_full_reviews["Price"] = scaler.fit_transform(df_full_reviews[["Price"]])
df_missing_reviews["Price"] = scaler.transform(df_missing_reviews[["Price"]])

# using best k-value to train model:
knn_final = KNeighborsRegressor(n_neighbors=k_star)
knn_final.fit(X, y)

# Normalizing missing data feature matrix:
X_missing = df_missing_reviews.loc[:, ["Price", "Beds"]]
X_missing = (X_missing - X.min()) / (X.max() - X.min())

# Predict missing values
y_missing_pred = knn_final.predict(X_missing)
df_missing_reviews["Review Scores Rating"] = y_missing_pred

# kde plot for observed vs predicted ratings:
plt.figure(figsize=(10, 6))
sns.kdeplot(y, label="Observed Ratings")
sns.kdeplot(y_missing_pred, label="Predicted Ratings")

# scaffolding:
plt.xlabel("Review Scores Rating")
plt.ylabel("Density")
plt.title("Kernel Density Plot of Observed vs Predicted Ratings")
plt.legend()
plt.show()
```

The predicted ratings have a much steeper curve compared to the observed ratings which is much more spread out. This is because the kNN method is predicting points based on the current mean which makes it so the points are much more clustered together compared to our observed data which tends to be more dispersed. A higher k value would result in more groups since the model is using more neighboring points to predict its values, resulting in a more underfit and general smoother curve.
